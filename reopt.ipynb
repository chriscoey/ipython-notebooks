## Models and Definitions

#### Stochastic Model

Let first and second stage costs be $c,d \geq 0$. Let uncertainty set $\mathcal{U} \in \mathbb{R}^m_+$ be defined by $\mathcal{U} = \{ h \in \mathbb{R}^m_+\ |\ Qh \leq r \}, Q \in \mathbb{R}^{k \times m}, r \in \mathbb{R}^{k}$, where $k$ is the number of constraints defining the polytope $\mathcal{U}$. Let constraint coefficients for both first and second stage decisions be $A \in \mathbb{R}^{m \times n}_+$. The stochastic model is:

$
Z^{stoch} = 
\min\ c'x + \mathbb{E}_h [d'y(h)] \\
\quad A (x + y(h)) \geq h \quad \forall h \in \mathcal{U} \\
\quad x, y(h) \in \mathbb{R}^n_+
$

#### SAA

Using SAA to approximate the stochastic objective, with samples $\mathcal{S}$ from $\mathcal{U}$:

$
Z^{SAA} = \min\ c'x + \frac{1}{|S|} \sum_{h \in S} d'y(h) \\
\quad A (x + y(h)) \geq h \quad \forall h \in \mathcal{S} \\
\quad x, y(h) \in \mathbb{R}^n_+
$

Then for large $|S|$, $Z^{SAA} \sim Z^{stoch}$.

#### Affine Decision Rule

Use an affine decision rule $\hat y(h) = Ph + q$ with $P \in \mathbb{R}^{n \times m}, q \in \mathbb{R}^n$. Let the first stage solutions for the stochastic and robust affine models be $x_{aff}^{stoch}$ and $x_{aff}^{rob}$. 

#### Stochastic Affine Model

The stochastic objective is $\min\ c'x_{aff}^{stoch} + \mathbb{E}_h [d'(Ph + q)] = c'x_{aff}^{stoch} + d'(P \bar h + q)$, where $\bar h = \mathbb{E}_h h$.

$
Z_{aff}^{stoch} = 
\min\ c'x_{aff}^{stoch} + d'P \bar h + d'q \\
\quad Ax_{aff}^{stoch} + APh + Aq \geq h \quad \forall h \in \mathcal{U} \\
\quad Ph + q \geq 0 \quad \forall h \in \mathcal{U} \\
\quad x_{aff}^{stoch} \in \mathbb{R}^n_+, q \in \mathbb{R}^n, P \in \mathbb{R}^{n \times m}
$

This problem is infinite constraints, so reformulate. For the first set of constraints, write as:

$
\min_h\ e'_i (Ax + APh + Aq) - h_i \geq 0 \quad \forall i \in 1 \dots m \\
\equiv e'_i Ax + e'_i Aq + \min_h\ e'_i (APh - h) \geq 0 \quad \forall i \in 1 \dots m
$

Consider $\min_h\ e'_i (APh - h) = \min_h\ e'_i (AP - I)h$. For $i \in 1 \dots m$, we have a primal problem in variable $h$:

$
\min\ e'_i (AP - I)h \\
\quad -Qh \geq -r \quad (\beta_i) \\
\quad h \geq 0
$

With dual:

$
\max\ -r' \beta_i \\
\quad -Q' \beta_i \leq (AP - I)' e_i \\
\quad \beta_i \geq 0
$

So, dropping the $\max$, the first set of constraints from the previous model can be written:

$
e'_i Ax + e'_i Aq - r' \beta_i \geq 0 \quad \forall i \in 1 \dots m \\
Q' \beta_i + (AP - I)'e_i \geq 0 \quad \forall i \in 1 \dots m \\
\beta_i \geq 0 \geq 0 \quad \forall i \in 1 \dots m 
$

The second set of constraints are equivalent to $\min_h\ e'_j Ph + e'_j q \geq 0\ \forall j \in 1 \dots n$. For some $j$, consider primal problem in variable $h$:

$
\min\ e'_j Ph \\
\quad -Qh \geq -r \quad (\gamma_j) \\
\quad h \geq 0
$

With dual:

$
\max\ -r' \gamma_j \\
\quad -Q' \gamma_j \leq P' e_j \\
\quad \gamma_j \geq 0
$

So, dropping the $\max$, the second set of constraints from the previous model can be written:

$
-r' \gamma_j + e'_j q \geq 0 \quad \forall j \in 1 \dots n \\
Q' \gamma_j + P' e_j \geq 0 \quad \forall j \in 1 \dots n \\
\gamma_j \geq 0 \quad \forall j \in 1 \dots n
$

The stochastic affine model with finite constraints is therefore:

$
Z_{aff}^{stoch} = 
\min\ c'x_{aff}^{stoch} + d'P \bar h + d'q \\
\quad e'_i Ax + e'_i Aq - r' \beta_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad Q' \beta_i + (AP - I)'e_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad -r' \gamma_j + e'_j q \geq 0 \quad \forall j \in 1 \dots n \\
\quad Q' \gamma_j + P' e_j \geq 0 \quad \forall j \in 1 \dots n \\
\quad x_{aff}^{stoch} \in \mathbb{R}^n_+, q \in \mathbb{R}^n, P \in \mathbb{R}^{n \times m}, \\
\quad \beta_i \in \mathbb{R}^k_+\ \forall i \in 1 \dots m, \gamma_j \in \mathbb{R}^k_+\ \forall j \in 1 \dots n
$

#### Robust Affine Model

The robust objective is $\min\ c'x + \max_h\ d'y(h)$. With an affine policy, the objective is $\min\ c'x_{aff}^{rob} + \max_h\ \{ d'(Ph + q) \}$. Reformulate the robust affine model to get a linear objective: $c'x_{aff}^{rob} + z$ where $z \in \mathbb{R}, z \geq (\max_h\ d'Ph) + d'q$. 

$
Z_{aff}^{rob} = 
\min\ c'x_{aff}^{rob} + z \\
\quad z \geq d'Ph + d'q \quad \forall h \in \mathcal{U} \\
\quad Ax_{aff}^{stoch} + APh + Aq \geq h \quad \forall h \in \mathcal{U} \\
\quad Ph + q \geq 0 \quad \forall h \in \mathcal{U} \\
\quad x_{aff}^{stoch} \in \mathbb{R}^n_+, q \in \mathbb{R}^n, P \in \mathbb{R}^{n \times m}, z \in \mathbb{R}
$

This problem is infinite constraints, so reformulate. Consider $\max_h\ d'Ph = \max_h\ (P'd)'h$. Since $\mathcal{U} = \{ h \in \mathbb{R}^m_+\ |\ Qh \leq r \}, Q \in \mathbb{R}^{k \times m}, r \in \mathbb{R}^{k}$, we have primal problem in variable $h$:

$
\max\ (P'd)'h \\
\quad Qh \leq r \quad (\alpha) \\
\quad h \geq 0
$

With dual:

$
\min\ r' \alpha \\
\quad Q' \alpha \geq P'd \quad (h) \\
\quad \alpha \geq 0
$

So, dropping the $\min$, the first set of constraints from the previous model can be written:

$
z \geq r' \alpha + d'q \\
Q' \alpha \geq P'd \\
\alpha \geq 0
$

For the second and third sets of constraints in the semi-infinite robust model, replace with the reformulated constraints derived for the stochastic model previously. The robust affine model with finite constraints is therefore:

$
Z_{aff}^{rob} = 
\min\ c'x_{aff}^{rob} + z \\
\quad z \geq r' \alpha + d'q \\
\quad Q' \alpha \geq P'd \\
\quad e'_i Ax + e'_i Aq - r' \beta_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad Q' \beta_i + (AP - I)'e_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad -r' \gamma_j + e'_j q \geq 0 \quad \forall j \in 1 \dots n \\
\quad Q' \gamma_j + P' e_j \geq 0 \quad \forall j \in 1 \dots n \\
\quad x_{aff}^{stoch} \in \mathbb{R}^n_+, q \in \mathbb{R}^n, P \in \mathbb{R}^{n \times m}, z \in \mathbb{R}, \\
\quad \alpha \in \mathbb{R}^k_+, \beta_i \in \mathbb{R}^k_+\ \forall i \in 1 \dots m, \gamma_j \in \mathbb{R}^k_+\ \forall j \in 1 \dots n
$

## Goals

Verify intuition: 
* $c \sim d \implies x_{aff}^{stoch}$ performs well i.e. $Z_H \sim \tilde Z_H(x_{aff}^{stoch})$
* $c \not\sim d \implies x_{aff}^{rob}$ performs well i.e. $Z_H \sim \tilde Z_H(x_{aff}^{rob})$




## Uncertainty Set

Let uncertainty set $\mathcal{U} \in \mathbb{R}^m_+$ be a uniform distribution over the $m$-dimensional polytope between the origin and standard simplex. Thus $\mathcal{U} = \{ h \in \mathbb{R}^m_+\ |\ \sum_{i=0}^m h_i \leq 1 \}$. Thus $k = 1$, since only one constraint defines the uncertainty set (except for nonnegativity constraints).

The mean of the uncertainty set is $\mathbb{E}_h [h] = \bar h$, and $\bar h_i = \frac{1}{m+1}\ \forall i \in 1 \dots m$. This is intuitive from the description of the sampling method below.

To create sample set $|S|$, a sample $h$ in $m$ dimensions can be generated by taking a random point from a Dirichlet distribution in $m+1$ dimensions and discarding one element of the vector.

Below is an example of an $m=2$ uncertainty set, plotted to visualize uniformity over the polytope. 

```julia
using Distributions
using Gadfly

D = Dirichlet([1.0,1.0,1.0])
D_s = rand(D,2000)
U_s = D_s[1:2,:]
plot(x=U_s[1,:], y=U_s[2,:], Geom.point, Theme(default_point_size=0.5mm))
```

Below we calculate the sample mean to verify that it is close to the true mean $\bar h = (\frac{1}{3}, \frac{1}{3})$.

```julia
E_hs = [mean(U_s[1,:]), mean(U_s[2,:])] #means of rows of samples E(h) over S
```

## Parameter Input

* $m =$ # dimensions of $\mathcal{U}$
* $K =$ # samples from $\mathcal{U}$, i.e. $|S|$
* $n =$ number of decisions $x, y$ in each stage
* $c =$ first stage costs (distribution params)
* $d =$ second stage costs (distribution params)
* $A =$ constraint coefficients for x and y (distribution params)

```julia
m = 10 #dimensions

K = 100 #sample size

n = 10 #decisions in each stage

c_mu = 1.0 #mean c
c_sig = 0.3 #variance c
d_eps = 0.1 #factor when c,d close
d_lf = 4 #low factor when c,d not close
d_hf = 8 #high factor when c,d not close

A_mu = 1/n #mean of entries in A
A_sig = 0.15 #SD of entries in A
```

## Data Generation

#### Uncertainty set

```julia
D = Dirichlet(fill(1.0,m+1))
D_s = rand(D,K)
U_s = D_s[1:m,:] #sample set S
E_hs = [mean(U_s[j,:]) for j=1:m] #means of rows of samples E(h)
hbar = 1/(m+1)
```

#### Costs

```julia
c = rand(Normal(c_mu, c_sig), n) #first stage costs vector
d_c = [rand(Uniform(c_j, (1+d_eps)*c_j)) for c_j in c] #c,d close - second stage costs vector
d_nc = [rand(Uniform(d_lf*c_j, d_hf*c_j)) for c_j in c] #c,d not close - second stage costs vector
using DataFrames #dataframe of costs
df = DataFrame(c=c, d_c=d_c, d_nc=d_nc)
#costs = [1:n c fill("c",n); 1:n d_c fill("d_c",n); 1:n d_nc fill("d_nc",n)]
#plot(x=costs[:,1], y=costs[:,2], color=costs[:,3]) #scatter plot of costs
```

#### Constraint Matrix

```julia
A = [max(rand(Normal(A_mu, A_sig)), 0) for _=1:m, _=1:n]
```

## Testing

#### SAA Solution

Find $Z^{SAA}$ for $c,d$ close ($Z^{SAA}_c$) and $c,d$ not close ($Z^{SAA}_{nc}$).

```julia
using JuMP
using Gurobi

function Find_Z_SAA(d) #d is d_c or d_nc
    mod = Model()
    @defVar(mod, x[1:n] >= 0)
    @defVar(mod, y[1:K,1:n] >= 0)
    @setObjective(mod, Min, sum{c[j]*x[j], j=1:n} + 1/K * sum{d[j]*y[k,j], j=1:n, k=1:K})
    for k=1:K
        for i=1:m
            @addConstraint(mod, sum{A[i,j]*x[j], j=1:n} + sum{A[i,j]*y[k,j], j=1:n} >= U_s[i,k])
        end
    end

    status = solve(mod)
    if status == :Infeasible
        error("Infeasible")
    end
    Z = getObjectiveValue(mod)
#    println()
#    println("Z_SAA(d) = $(Z)")
#    println()
#    println("x =")
#    println(getValue(x))
#    println()
#    println("y =")
#    println(getValue(y))
    return Z
end

Z_SAA_c = Find_Z_SAA(d_c)
Z_SAA_nc = Find_Z_SAA(d_nc)
[Z_SAA_c, Z_SAA_nc]
```

#### Stochastic Affine Solution

Find $Z_{aff}^{stoch}$ for $c,d$ close ($Z_{aff, c}^{stoch}$) and $c,d$ not close ($Z_{aff, nc}^{stoch}$).

Since we are using the unit simplex uncertainty set form $\mathcal{U} = \{ h \in \mathbb{R}^m_+\ |\ \sum_{i=0}^m h_i \leq 1 \}$, we have $k = 1$, $Q = \mathbf{1}$ and $r = 1$. We have $\bar h_i = \frac{1}{m+1}\ \forall i \in 1 \dots m$.

$
Z_{aff}^{stoch} = 
\min\ \sum_{j=1}^n c_j x_{aff, j}^{stoch} + \frac{1}{m+1} \sum_{j=1}^n d_j \sum_{i=1}^m P_{j,i} + \sum_{j=1}^n d_j q_j \\
\quad \sum_{j=1}^n A_{i,j} x_j + \sum_{j=1}^n A_{i,j} q_j - \beta_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad \beta_i + \sum_{c = 1}^n A_{l,c} P_{c,i} \geq 0 \quad \forall i \in 1 \dots m, l \in 1 \dots m, i \neq l \\
\quad \beta_i + \sum_{c = 1}^n A_{i,c} P_{c,i} - 1 \geq 0 \quad \forall i \in 1 \dots m \\
\quad \gamma_j + q_j \geq 0 \quad \forall j \in 1 \dots n \\
\quad \gamma_j + P_{i,j} \geq 0 \quad \forall j \in 1 \dots n, i \in 1 \dots m \\
\quad q, P \text{ free} \\
\quad x_{aff}^{stoch} \geq 0 \\
\quad \beta_i \geq 0\ \forall i \in 1 \dots m \\
\quad \gamma_j \geq 0\ \forall j \in 1 \dots n
$

# if either P or q are free, solution is unbounded. if both nonneg, bounded

```julia
function Find_x_aff_stoch(d) #d is d_c or d_nc

    mod = Model()

    @defVar(mod, x[1:n] >= 0)
    @defVar(mod, q[1:n] >= 0)
    @defVar(mod, P[1:n,1:m] >= 0)
    @defVar(mod, b[1:m] >= 0)
    @defVar(mod, g[1:n] >= 0)

    @setObjective(mod, Min, sum{c[j]*x[j], j=1:n} + sum{d[j]*hbar*P[j,i],
        i=1:m, j=1:n} + sum{d[j]*q[j], j=1:n})
    
    for i=1:m
        @addConstraint(mod, sum{A[i,j]*x[j], j=1:n} + sum{A[i,j]*q[j], j=1:n} - b[i] >= 0)
        for l=1:m
            if i==l
                @addConstraint(mod, b[i] + sum{A[l,c]*P[c,i], c=1:n} - 1 >= 0)
            else
                @addConstraint(mod, b[i] + sum{A[l,c]*P[c,i], c=1:n} >= 0)
            end
        end
        for j=1:n
            @addConstraint(mod, g[j] + P[i,j] >= 0)   
        end
    end
    for j=1:n
        @addConstraint(mod, g[j] + q[j] >= 0)
    end

    status = solve(mod)
    if status == :Infeasible
        error("Infeasible")
    end
    Z = getObjectiveValue(mod)
#    println()
#    println("Z_aff_stoch(d) = $(Z)")
#    println()
#    println("x =")
#    println(getValue(x))
#    println()
#    println("y =")
#    println(getValue(y))
    return Z
end

Z_x_aff_stoch_c = Find_x_aff_stoch(d_c)
Z_x_aff_stoch_nc = Find_x_aff_stoch(d_nc)
[Z_x_aff_stoch_c, Z_x_aff_stoch_nc]
```

#### Robust Affine Solution

Find $Z_{aff}^{rob}$ for $c,d$ close ($Z_{aff, c}^{rob}$) and $c,d$ not close ($Z_{aff, nc}^{rob}$).

Since we are using the unit simplex uncertainty set form $\mathcal{U} = \{ h \in \mathbb{R}^m_+\ |\ \sum_{i=0}^m h_i \leq 1 \}$, we have $k = 1$, $Q = \mathbf{1}$ and $r = 1$. We have $\bar h_i = \frac{1}{m+1}\ \forall i \in 1 \dots m$.

$
Z_{aff}^{rob} = 
\min\ \sum_{j=1}^n c_j x_{aff, j}^{rob} + z \\
\quad z - \alpha - \sum_{j=1}^n d_j q_j \geq 0 \\
\quad \alpha - \sum_{j=1}^n d_j P_{j,i} \geq 0 \quad \forall i \in 1 \dots m \\
\quad \sum_{j=1}^n A_{i,j} x_j + \sum_{j=1}^n A_{i,j} q_j - \beta_i \geq 0 \quad \forall i \in 1 \dots m \\
\quad \beta_i + \sum_{c = 1}^n A_{l,c} P_{c,i} \geq 0 \quad \forall i \in 1 \dots m, l \in 1 \dots m, i \neq l \\
\quad \beta_i + \sum_{c = 1}^n A_{i,c} P_{c,i} - 1 \geq 0 \quad \forall i \in 1 \dots m \\
\quad \gamma_j + q_j \geq 0 \quad \forall j \in 1 \dots n \\
\quad \gamma_j + P_{i,j} \geq 0 \quad \forall j \in 1 \dots n, i \in 1 \dots m \\
\quad z, q, P \text{ free} \\
\quad x_{aff}^{stoch}, \alpha \geq 0 \\
\quad \beta_i \geq 0\ \forall i \in 1 \dots m \\
\quad \gamma_j \geq 0\ \forall j \in 1 \dots n
$

# if either P or q are free, solution is unbounded. if both nonneg, bounded

```julia
function Find_x_aff_rob(d) #d is d_c or d_nc

    mod = Model()

    @defVar(mod, x[1:n] >= 0)
    @defVar(mod, z)
    @defVar(mod, q[1:n] >= 0)
    @defVar(mod, P[1:n,1:m] >= 0)
    @defVar(mod, a >= 0)
    @defVar(mod, b[1:m] >= 0)
    @defVar(mod, g[1:n] >= 0)

    @setObjective(mod, Min, sum{c[j]*x[j], j=1:n} + z)
    
    @addConstraint(mod, z - a - sum{d[j]*q[j], j=1:n} >= 0)
    for i=1:m
        @addConstraint(mod, a - sum{d[j]*P[j,i], j=1:n} >= 0)
        @addConstraint(mod, sum{A[i,j]*x[j], j=1:n} + sum{A[i,j]*q[j], j=1:n} - b[i] >= 0)
        for l=1:m
            if i==l
                @addConstraint(mod, b[i] + sum{A[l,c]*P[c,i], c=1:n} - 1 >= 0)
            else
                @addConstraint(mod, b[i] + sum{A[l,c]*P[c,i], c=1:n} >= 0)
            end
        end
        for j=1:n
            @addConstraint(mod, g[j] + P[i,j] >= 0)   
        end
    end
    for j=1:n
        @addConstraint(mod, g[j] + q[j] >= 0)
    end

    status = solve(mod)
    if status == :Infeasible
        error("Infeasible")
    end
    Z = getObjectiveValue(mod)
#    println()
#    println("Z_aff_rob(d) = $(Z)")
#    println()
#    println(getValue(x))
#    println()
#    println(getValue(y))
end

Z_x_aff_rob_c = Find_x_aff_rob(d_c)
Z_x_aff_rob_nc = Find_x_aff_rob(d_nc)
[Z_x_aff_rob_c, Z_x_aff_rob_nc]
```
